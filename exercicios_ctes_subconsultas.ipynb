```python
def consulta(query, tabela):
  connection = mysql.connector.connect(
      host = host,
      user = user,
      password = password,
      database = database
  )
  cursor = connection.cursor()

  try:
    cursor.execute(query)
    result = cursor.fetchall()
    globals()[tabela] = pd.DataFrame(result, columns=cursor.column_names)
    display(globals()[tabela])
  finally:
    cursor.close()
    connection.close()
```

```python
def ddl(comando):
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )
    cursor = connection.cursor()

    try:
        cursor.execute(comando)
        connection.commit()  # Confirma as mudanças no banco
        print(f"Comando DDL executado com sucesso:\n{comando}")
    except mysql.connector.Error as err:
        print(f"Erro ao executar o comando DDL: {err}")
    finally:
        cursor.close()
        connection.close()
```

### 3. Liste os nomes dos clientes que alugaram pelo menos um carro da marca Ford.
Dica: Use uma subconsulta para filtrar a marca.

```python
'''
Liste os nomes dos clientes que alugaram pelo menos um carro da marca Ford.
Dica: Use uma subconsulta para filtrar a marca.

SELECT DISTINCT nome
FROM cliente
WHERE codcliente IN (
    SELECT codcliente
    FROM aluguel
    WHERE codcarro IN (
        SELECT codcarro
        FROM carro
        WHERE codmarca = (
            SELECT codmarca
            FROM marca
            WHERE marca = 'Ford'
        )
    )
);
'''
consulta()
```

### 4. Carros alugados apenas por clientes de São Paulo Liste os modelos de carros que foram alugados exclusivamente por clientes do estado de São Paulo.
Dica: Use NOT IN com uma subconsulta.


```python
'''
SELECT DISTINCT c.modelo
FROM carro c
JOIN aluguel a ON c.codcarro = a.codcarro
JOIN cliente cl ON a.codcliente = cl.codcliente
WHERE c.codcarro NOT IN (
    SELECT a2.codcarro
    FROM aluguel a2
    JOIN cliente cl2 ON a2.codcliente = cl2.codcliente
    WHERE cl2.estado <> 'SP'
);
'''
consulta()
```

### 5. Datas mais movimentadas Liste as datas em que houve mais de 2 aluguéis.
Dica: Use uma subconsulta com GROUP BY e HAVING.

```python
'''
SELECT DISTINCT data_aluguel
FROM aluguel
WHERE data_aluguel IN (
    SELECT data_aluguel
    FROM aluguel
    GROUP BY data_aluguel
    HAVING COUNT(*) > 2
);
'''
consulta()
```

### 7. Liste os modelos de carros que foram alugados em cidades onde residem mais de 3 clientes.
Dica: Use uma subconsulta para encontrar essas cidades.


```python
'''
SELECT DISTINCT c.modelo
FROM carro c
JOIN aluguel a ON c.codcarro = a.codcarro
JOIN cliente cl ON a.codcliente = cl.codcliente
JOIN (
    SELECT cidade
    FROM cliente
    GROUP BY cidade
    HAVING COUNT(*) > 3
) AS cidades_com_mais_de_3_clientes ON cl.cidade = cidades_com_mais_de_3_clientes.cidade;

'''
consulta()
```

### 8. Liste os nomes dos clientes e o valor total gasto em aluguéis.

### 9. Liste os nomes dos clientes cujo valor total gasto com aluguéis está acima da média geral.


```python
'''
SELECT cl.nome
FROM cliente cl
JOIN aluguel a ON cl.codcliente = a.codcliente
JOIN carro c ON a.codcarro = c.codcarro
GROUP BY cl.codcliente, cl.nome
HAVING SUM(c.valor) > (
    SELECT AVG(valor_total)
    FROM (
        SELECT SUM(c2.valor) AS valor_total
        FROM aluguel a2
        JOIN carro c2 ON a2.codcarro = c2.codcarro
        GROUP BY a2.codcliente
    ) AS totais_clientes
);
'''
consulta()
```

### 10. Liste os nomes dos clientes que realizaram o maior número de aluguéis em um único dia.
Dica: Use MAX em uma subconsulta.


```python
'''
SELECT
    cl.nome,
    apd.data_aluguel,
    apd.qtd_alugueis
FROM cliente cl
JOIN (
    SELECT codcliente, data_aluguel, COUNT(*) AS qtd_alugueis
    FROM aluguel
    GROUP BY codcliente, data_aluguel
) AS apd ON cl.codcliente = apd.codcliente
WHERE apd.qtd_alugueis = (
    SELECT MAX(contador)
    FROM (
        SELECT COUNT(*) AS contador
        FROM aluguel
        GROUP BY codcliente, data_aluguel
    ) AS sub_max
);
'''
consulta()
```

### 11. Liste os nomes dos clientes que alugaram todos os carros disponíveis da marca Chevrolet.
Dica: Compare a contagem de carros alugados por cliente com a contagem total de carros da marca.


```python
'''
SELECT cl.nome
FROM cliente cl
WHERE cl.codcliente IN (
    SELECT a.codcliente
    FROM aluguel a
    JOIN carro c ON a.codcarro = c.codcarro
    WHERE c.codmarca = (
        SELECT codmarca
        FROM marca
        WHERE marca = 'Chevrolet'
    )
    GROUP BY a.codcliente
    HAVING COUNT(DISTINCT a.codcarro) = (
        SELECT COUNT(*)
        FROM carro
        WHERE codmarca = (
            SELECT codmarca
            FROM marca
            WHERE marca = 'Chevrolet'
        )
    )
);
'''
consulta()
```

### 16. Crie uma CTE para calcular o número total de aluguéis realizados por cada cliente e liste os clientes que alugaram mais de 2 vezes.

```python
'''
WITH total_alugueis_por_cliente AS (
    SELECT codcliente, COUNT(*) AS total_alugueis
    FROM aluguel
    GROUP BY codcliente
)
SELECT c.nome, t.total_alugueis
FROM total_alugueis_por_cliente t
JOIN cliente c ON t.codcliente = c.codcliente
WHERE t.total_alugueis > 2;
'''
consulta()
```

### 17. Crie uma CTE para calcular o valor total gasto por cliente e liste apenas os clientes que gastaram mais de R$300,00.


```python
'''
WITH total_gasto_por_cliente AS (
    SELECT a.codcliente, SUM(c.valor) AS total_gasto
    FROM aluguel a
    JOIN carro c ON a.codcarro = c.codcarro
    GROUP BY a.codcliente
)
SELECT cl.nome, tg.total_gasto
FROM total_gasto_por_cliente tg
JOIN cliente cl ON tg.codcliente = cl.codcliente
WHERE tg.total_gasto > 300;
'''
consulta()
```

### 18. Crie uma CTE para calcular o número de vezes que cada modelo de carro foi alugado e liste apenas os modelos alugados mais de 3 vezes.


```python
'''
WITH alugueis_por_modelo AS (
    SELECT c.modelo, COUNT(*) AS total_alugueis
    FROM aluguel a
    JOIN carro c ON a.codcarro = c.codcarro
    GROUP BY c.modelo
)
SELECT modelo, total_alugueis
FROM alugueis_por_modelo
WHERE total_alugueis > 3;
'''
consulta()
```

### 19. Crie uma CTE para calcular o número de clientes por estado e liste os estados com mais de 3 clientes.


```python
'''
WITH clientes_por_estado AS (
    SELECT estado, COUNT(*) AS total_clientes
    FROM cliente
    GROUP BY estado
)
SELECT estado, total_clientes
FROM clientes_por_estado
WHERE total_clientes > 3;
'''
consulta()
```

### 20. Crie uma CTE para listar os clientes que realizaram aluguéis em dias consecutivos.
Dica: Use a função DATEDIFF.


```python
'''
WITH alugueis_com_datas AS (
    SELECT codcliente, data_aluguel
    FROM aluguel
)
SELECT DISTINCT cl.nome
FROM alugueis_com_datas a1
JOIN alugueis_com_datas a2
  ON a1.codcliente = a2.codcliente
  AND DATEDIFF(a1.data_aluguel, a2.data_aluguel) = 1
JOIN cliente cl ON cl.codcliente = a1.codcliente;
'''
consulta()
```
